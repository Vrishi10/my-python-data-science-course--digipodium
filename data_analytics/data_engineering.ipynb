```python
df.rename(mapper={"make":"brand"}, axis=1, inplace=True)          #axis is used to scan the column as it scans rowwise
```

```python
df.replace('?', np.nan, inplace=True)         #removes all ? byg np.nan
```

```python
df.dropna(subset=['price'])       #dropping of the record whose price is missing
```

```python
df.isnull().sum()    #counts the number of missing data in each column
```

```python
df['normalized-losses'].astype('float').sum()
```

```python
df.isnull().sum()     #counts the number of data in missing column
```

```python
mnl = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan, mnl, inplace=True)
df.head()
```

```python
df['bore'].replace(np.nan, df['bore'].astype('float').mean(), inplace=True)
df['stroke'].replace(np.nan, df['stroke'].astype('float').mean(), inplace=True)
df['horsepower'].replace(np.nan, df['horsepower'].astype('float').mean(), inplace=True)
df['peak-rpm'].replace(np.nan, df['peak-rpm'].astype('float').mean(), inplace=True)
df['price'].replace(np.nan, df['price'].astype('float').mean(), inplace=True)
df.head()

#in place of nan fill the mean value
```
